# Fair by Force on FLAIR Server (2-Hour Optimized)

This notebook is implementation-ready for federated fairness experiments with:

- Two run flags: `DRY_RUN` and `FULL_RUN`
- Highly verbose real-time logging (step/round/epoch/batch)
- Round-level checkpointing + resume
- Runtime budget guardrail (hard stop before 2 hours)
- Adaptive ETA estimator with uncertainty band
- JSON/CSV result export

## How to use

1. Run all setup cells.
2. In the configuration section, set exactly one of `DRY_RUN=True` or `FULL_RUN=True`.
3. Run the execution section.
4. Monitor live logs and artifacts in `artifacts/`.

## 2) Define Configuration and Runtime Parameters

Set all runtime knobs, enforce the two-flag mode, and tune defaults for a maximum 2-hour runtime.

```python
# ===== Required run flags (set exactly one True) =====
DRY_RUN = True
FULL_RUN = False
RESUME_FROM_CHECKPOINT = True

if DRY_RUN == FULL_RUN:
    raise ValueError("Exactly one of DRY_RUN or FULL_RUN must be True.")


@dataclass
class ExperimentConfig:
    dry_run: bool
    full_run: bool
    resume_from_checkpoint: bool = True

    # Runtime budget: hard stop before 2h (default: 1h55m)
    runtime_budget_hours: float = 2.0
    runtime_buffer_minutes: int = 5

    # Paths
    data_root: Path = Path("data")
    artifacts_root: Path = Path("artifacts")
    checkpoints_root: Path = Path("artifacts/checkpoints")
    logs_root: Path = Path("artifacts/logs")
    results_root: Path = Path("artifacts/results")

    # Data
    image_size: int = 128
    val_split: float = 0.15
    num_workers: int = 4
    pin_memory: bool = True
    strict_real_data: bool = False

    # Federated settings (2h optimized)
    num_clients: int = 6
    clients_per_round: int = 3
    rounds: int = 30
    local_epochs: int = 1
    batch_size: int = 64
    max_batches_per_epoch: int = 40

    # Optimization
    lr: float = 1e-3
    weight_decay: float = 1e-4
    use_amp: bool = True

    # Logging / checkpointing
    checkpoint_every_round: int = 1
    log_batch_every: int = 5

    # Fairness experiment conditions
    condition_names: List[str] = field(default_factory=lambda: ["vanilla", "A1_only"])

    # Dry-run overrides
    dry_run_rounds: int = 2
    dry_run_clients: int = 3
    dry_run_clients_per_round: int = 3
    dry_run_batch_size: int = 16
    dry_run_max_batches_per_epoch: int = 4

    def runtime_budget_seconds(self) -> int:
        safe_minutes = max(0, self.runtime_buffer_minutes)
        return int(self.runtime_budget_hours * 3600) - safe_minutes * 60


def build_config(dry_run: bool, full_run: bool, resume_from_checkpoint: bool) -> ExperimentConfig:
    cfg = ExperimentConfig(dry_run=dry_run, full_run=full_run, resume_from_checkpoint=resume_from_checkpoint)

    if cfg.dry_run:
        cfg.rounds = cfg.dry_run_rounds
        cfg.num_clients = cfg.dry_run_clients
        cfg.clients_per_round = cfg.dry_run_clients_per_round
        cfg.batch_size = cfg.dry_run_batch_size
        cfg.max_batches_per_epoch = cfg.dry_run_max_batches_per_epoch
        cfg.condition_names = ["vanilla", "A1_only"]

    if cfg.clients_per_round > cfg.num_clients:
        raise ValueError("clients_per_round cannot be greater than num_clients")

    if cfg.runtime_budget_seconds() <= 0:
        raise ValueError("Runtime budget must be positive after buffer subtraction")

    cfg.artifacts_root.mkdir(parents=True, exist_ok=True)
    cfg.checkpoints_root.mkdir(parents=True, exist_ok=True)
    cfg.logs_root.mkdir(parents=True, exist_ok=True)
    cfg.results_root.mkdir(parents=True, exist_ok=True)

    return cfg


CFG = build_config(DRY_RUN, FULL_RUN, RESUME_FROM_CHECKPOINT)
print("Config ready:")
print(json.dumps({k: (str(v) if isinstance(v, Path) else v) for k, v in asdict(CFG).items()}, indent=2))
print(f"Hard runtime budget (seconds): {CFG.runtime_budget_seconds()}")
```

## 4) Write the First Working Functions

Implement deterministic setup, verbose logger, adaptive ETA estimator, fairness metrics, and checkpoint primitives.

```python
def set_all_seeds(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)


def get_available_devices() -> List[torch.device]:
    if torch.cuda.is_available():
        return [torch.device(f"cuda:{i}") for i in range(torch.cuda.device_count())]
    return [torch.device("cpu")]


def format_seconds(sec: Optional[float]) -> str:
    if sec is None or math.isnan(sec):
        return "unknown"
    sec = max(0.0, float(sec))
    h = int(sec // 3600)
    m = int((sec % 3600) // 60)
    s = int(sec % 60)
    return f"{h:02d}:{m:02d}:{s:02d}"


class AdaptiveETAEstimator:
    def __init__(self, alpha: float = 0.30):
        self.alpha = alpha
        self.ewma: Optional[float] = None
        self.history: List[float] = []

    def update(self, duration_sec: float) -> None:
        duration_sec = float(max(0.0, duration_sec))
        self.history.append(duration_sec)
        if self.ewma is None:
            self.ewma = duration_sec
        else:
            self.ewma = self.alpha * duration_sec + (1.0 - self.alpha) * self.ewma

    def estimate(self, remaining_steps: int) -> Tuple[float, float, float]:
        if remaining_steps <= 0:
            return 0.0, 0.0, 0.0

        if not self.history:
            return float("nan"), float("nan"), float("nan")

        base = self.ewma if self.ewma is not None else float(np.mean(self.history))
        median = float(np.median(self.history))
        robust_step = 0.7 * base + 0.3 * median
        eta = robust_step * remaining_steps

        if len(self.history) >= 4:
            q25, q75 = np.quantile(np.array(self.history), [0.25, 0.75])
            eta_low = float(max(0.0, q25 * remaining_steps))
            eta_high = float(max(0.0, q75 * remaining_steps))
        else:
            eta_low = 0.8 * eta
            eta_high = 1.2 * eta

        return float(eta), float(eta_low), float(eta_high)


def build_run_logger(log_root: Path, run_name: str) -> logging.Logger:
    log_root.mkdir(parents=True, exist_ok=True)
    logger = logging.getLogger(f"flair_run_{run_name}")
    logger.setLevel(logging.INFO)
    logger.handlers.clear()

    formatter = logging.Formatter("%(asctime)s | %(levelname)s | %(message)s")

    stream_handler = logging.StreamHandler(sys.stdout)
    stream_handler.setFormatter(formatter)
    logger.addHandler(stream_handler)

    file_path = log_root / f"{run_name}.log"
    file_handler = logging.FileHandler(file_path, mode="a", encoding="utf-8")
    file_handler.setFormatter(formatter)
    logger.addHandler(file_handler)

    logger.propagate = False
    logger.info(f"Logger initialized. File: {file_path}")
    return logger


def _safe_rate(mask: np.ndarray, preds: np.ndarray) -> float:
    if mask.sum() == 0:
        return 0.0
    return float(preds[mask].mean())


def demographic_parity_gap(preds: np.ndarray, protected: np.ndarray) -> float:
    group0 = protected == 0
    group1 = protected == 1
    rate0 = _safe_rate(group0, preds)
    rate1 = _safe_rate(group1, preds)
    return abs(rate0 - rate1)


def equalized_odds_gap(preds: np.ndarray, labels: np.ndarray, protected: np.ndarray) -> float:
    worst = 0.0
    for y in [0, 1]:
        y_mask = labels == y
        g0 = np.logical_and(y_mask, protected == 0)
        g1 = np.logical_and(y_mask, protected == 1)
        r0 = _safe_rate(g0, preds)
        r1 = _safe_rate(g1, preds)
        worst = max(worst, abs(r0 - r1))
    return float(worst)


def accuracy_from_logits(logits: torch.Tensor, labels: torch.Tensor) -> float:
    pred = torch.argmax(logits, dim=1)
    return float((pred == labels).float().mean().item())


class CheckpointManager:
    def __init__(self, checkpoints_root: Path, condition_name: str, logger: logging.Logger):
        self.condition_name = condition_name
        self.logger = logger
        self.dir = checkpoints_root / condition_name
        self.dir.mkdir(parents=True, exist_ok=True)
        self.metrics_jsonl = self.dir / "round_metrics.jsonl"

    def checkpoint_path(self, round_idx: int) -> Path:
        return self.dir / f"round_{round_idx:04d}.pt"

    def _round_from_name(self, path: Path) -> int:
        m = re.search(r"round_(\d+)\.pt", path.name)
        return int(m.group(1)) if m else -1

    def latest_checkpoint(self) -> Optional[Path]:
        candidates = sorted(self.dir.glob("round_*.pt"), key=self._round_from_name)
        return candidates[-1] if candidates else None

    def save(
        self,
        round_idx: int,
        condition_name: str,
        global_model_state: Dict[str, torch.Tensor],
        history: List[Dict[str, Any]],
        wallclock_elapsed_sec: float,
    ) -> Path:
        payload = {
            "condition_name": condition_name,
            "round_idx": round_idx,
            "global_model_state": {k: v.detach().cpu() for k, v in global_model_state.items()},
            "history": history,
            "wallclock_elapsed_sec": float(wallclock_elapsed_sec),
            "saved_at": datetime.now(timezone.utc).isoformat(),
        }
        path = self.checkpoint_path(round_idx)
        torch.save(payload, path)
        self.logger.info(f"[checkpoint] saved -> {path}")
        return path

    def append_round_metrics(self, metrics: Dict[str, Any]) -> None:
        with self.metrics_jsonl.open("a", encoding="utf-8") as f:
            f.write(json.dumps(metrics) + "\n")

    def load_latest(self) -> Optional[CheckpointState]:
        latest = self.latest_checkpoint()
        if latest is None:
            return None
        payload = torch.load(latest, map_location="cpu")
        self.logger.info(f"[checkpoint] loaded <- {latest}")
        return CheckpointState(
            condition_name=payload["condition_name"],
            round_idx=int(payload["round_idx"]),
            global_model_state=payload["global_model_state"],
            history=payload.get("history", []),
            wallclock_elapsed_sec=float(payload.get("wallclock_elapsed_sec", 0.0)),
        )


print("Core utility functions ready.")
```

## 5) Run an End-to-End Execution Cell

Implement dataset/model/training pipeline, then execute a complete dry run or full run based on the run flags.

```python
class UTKFaceGenderDataset(Dataset):
    """UTKFace dataset parser using filename pattern: age_gender_race_date.jpg."""

    def __init__(self, image_paths: List[Path], transform: transforms.Compose):
        self.transform = transform
        self.samples: List[Tuple[Path, int, int]] = []

        for p in image_paths:
            parts = p.stem.split("_")
            if len(parts) < 3:
                continue
            try:
                gender = int(parts[1])
                race = int(parts[2])
            except ValueError:
                continue

            # Binary protected attribute: {0,1} where 0 means race in {0,1}, 1 means otherwise.
            protected = 0 if race in (0, 1) else 1
            label = 1 if gender == 1 else 0
            self.samples.append((p, label, protected))

        if not self.samples:
            raise RuntimeError("No valid UTKFace samples found after parsing filenames")

    def __len__(self) -> int:
        return len(self.samples)

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
        path, label, protected = self.samples[idx]
        img = Image.open(path).convert("RGB")
        img = self.transform(img)
        return img, torch.tensor(label, dtype=torch.long), torch.tensor(protected, dtype=torch.long)


class SyntheticFairDataset(Dataset):
    """Synthetic fallback dataset for dry-run and functional validation."""

    def __init__(self, size: int, image_size: int, seed: int):
        transform = transforms.Compose(
            [
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ]
        )
        self.base = FakeData(
            size=size,
            image_size=(3, image_size, image_size),
            num_classes=2,
            transform=transform,
            random_offset=seed,
        )
        rng = np.random.default_rng(seed)
        self.protected = rng.integers(low=0, high=2, size=size, endpoint=False)

    def __len__(self) -> int:
        return len(self.base)

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
        x, y = self.base[idx]
        s = int(self.protected[idx])
        return x, torch.tensor(int(y), dtype=torch.long), torch.tensor(s, dtype=torch.long)


def build_transforms(image_size: int) -> transforms.Compose:
    return transforms.Compose(
        [
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ]
    )


def discover_utkface_images(data_root: Path) -> List[Path]:
    candidates = [
        data_root / "UTKFace",
        data_root / "utkface",
        data_root,
        Path("UTKFace"),
    ]
    image_paths: List[Path] = []
    exts = ["*.jpg", "*.jpeg", "*.png"]

    for c in candidates:
        if c.exists():
            for ext in exts:
                image_paths.extend(list(c.rglob(ext)))
        if image_paths:
            break

    return sorted(image_paths)


def make_iid_partitions(indices: List[int], num_clients: int, seed: int) -> Dict[int, List[int]]:
    rng = np.random.default_rng(seed)
    shuffled = np.array(indices)
    rng.shuffle(shuffled)
    splits = np.array_split(shuffled, num_clients)
    return {cid: split.astype(int).tolist() for cid, split in enumerate(splits)}


def build_datasets_and_partitions(cfg: ExperimentConfig, logger: logging.Logger) -> Tuple[Dataset, Dataset, Dict[int, List[int]], str]:
    transform = build_transforms(cfg.image_size)
    image_paths = discover_utkface_images(cfg.data_root)

    if image_paths:
        logger.info(f"Found {len(image_paths)} image files. Attempting UTKFace parsing...")
        try:
            full_dataset = UTKFaceGenderDataset(image_paths, transform=transform)
            source = "utkface"
            logger.info(f"UTKFace loaded with {len(full_dataset)} valid samples.")
        except Exception as ex:
            if cfg.strict_real_data:
                raise
            logger.warning(f"UTKFace parse failed ({ex}); using synthetic fallback dataset.")
            full_dataset = SyntheticFairDataset(size=4000 if cfg.full_run else 800, image_size=cfg.image_size, seed=SEED)
            source = "synthetic_fallback"
    else:
        if cfg.strict_real_data:
            raise FileNotFoundError("No UTKFace images found and strict_real_data=True")
        logger.warning("UTKFace not found; using synthetic fallback dataset.")
        full_dataset = SyntheticFairDataset(size=4000 if cfg.full_run else 800, image_size=cfg.image_size, seed=SEED)
        source = "synthetic_fallback"

    n = len(full_dataset)
    val_n = max(1, int(n * cfg.val_split))
    rng = np.random.default_rng(SEED)
    all_indices = np.arange(n)
    rng.shuffle(all_indices)
    val_indices = all_indices[:val_n].tolist()
    train_indices = all_indices[val_n:].tolist()

    client_partitions = make_iid_partitions(train_indices, cfg.num_clients, seed=SEED)
    train_dataset = full_dataset
    val_dataset = Subset(full_dataset, val_indices)

    logger.info(
        f"Dataset source={source} | total={n} | train={len(train_indices)} | val={len(val_indices)} | clients={cfg.num_clients}"
    )
    return train_dataset, val_dataset, client_partitions, source


def build_model(num_classes: int, device: torch.device, logger: logging.Logger) -> nn.Module:
    try:
        weights = models.ResNet18_Weights.DEFAULT
        model = models.resnet18(weights=weights)
        logger.info("Loaded ResNet18 with pretrained ImageNet weights.")
    except Exception as ex:
        logger.warning(f"Pretrained weights unavailable ({ex}); using random initialization.")
        model = models.resnet18(weights=None)

    in_features = model.fc.in_features
    model.fc = nn.Linear(in_features, num_classes)
    return model.to(device)


def apply_attack_to_loss(per_sample_loss: torch.Tensor, protected: torch.Tensor, attack_name: str) -> torch.Tensor:
    if attack_name == "none":
        return per_sample_loss.mean()

    if attack_name == "A1":
        # Inverse-frequency reweighting by protected group.
        groups, counts = torch.unique(protected, return_counts=True)
        inv_freq = {int(g.item()): 1.0 / max(1, int(c.item())) for g, c in zip(groups, counts)}
        weights = torch.tensor([inv_freq[int(g.item())] for g in protected], device=per_sample_loss.device)
        weights = weights / weights.mean().clamp(min=1e-8)
        return (per_sample_loss * weights).mean()

    # A2-A5 placeholders: extend with full paper-specific operators as needed.
    if attack_name in {"A2", "A3", "A4", "A5"}:
        return per_sample_loss.mean()

    raise ValueError(f"Unknown attack_name={attack_name}")


def apply_defense_to_gradients(model: nn.Module, defense_name: str) -> float:
    grad_norm_sq = 0.0
    for p in model.parameters():
        if p.grad is not None:
            grad_norm_sq += float(torch.sum(p.grad.detach() ** 2).item())
    grad_norm = math.sqrt(max(0.0, grad_norm_sq))

    if defense_name == "none":
        return grad_norm
    if defense_name == "D1":
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=2.0)
    elif defense_name == "D2":
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    elif defense_name == "D3":
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
    elif defense_name in {"D4", "D5"}:
        # Representation-level defense hooks can be inserted here.
        pass
    else:
        raise ValueError(f"Unknown defense_name={defense_name}")

    return grad_norm


def local_train_one_client(
    client_id: int,
    device: torch.device,
    global_state: Dict[str, torch.Tensor],
    train_dataset: Dataset,
    sample_indices: List[int],
    cfg: ExperimentConfig,
    condition: AttackDefenseCondition,
    logger: logging.Logger,
) -> Tuple[Dict[str, torch.Tensor], ClientMetrics]:
    subset = Subset(train_dataset, sample_indices)
    loader = DataLoader(
        subset,
        batch_size=cfg.batch_size,
        shuffle=True,
        num_workers=cfg.num_workers,
        pin_memory=(cfg.pin_memory and device.type == "cuda"),
        drop_last=False,
    )

    model = build_model(num_classes=2, device=device, logger=logger)
    model.load_state_dict(global_state)
    model.train()

    optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
    criterion = nn.CrossEntropyLoss(reduction="none")

    use_amp = bool(cfg.use_amp and device.type == "cuda")
    scaler = torch.cuda.amp.GradScaler(enabled=use_amp)

    all_preds: List[np.ndarray] = []
    all_labels: List[np.ndarray] = []
    all_protected: List[np.ndarray] = []
    losses: List[float] = []
    accs: List[float] = []

    t0 = time.perf_counter()
    for epoch in range(1, cfg.local_epochs + 1):
        logger.info(f"[client={client_id}] epoch={epoch}/{cfg.local_epochs} start | batches={len(loader)}")

        for batch_idx, (x, y, s) in enumerate(loader, start=1):
            if batch_idx > cfg.max_batches_per_epoch:
                logger.info(f"[client={client_id}] max_batches_per_epoch reached ({cfg.max_batches_per_epoch}).")
                break

            x = x.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)
            s = s.to(device, non_blocking=True)

            optimizer.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=use_amp):
                logits = model(x)
                per_sample_loss = criterion(logits, y)
                loss = apply_attack_to_loss(per_sample_loss, s, condition.attack)

            scaler.scale(loss).backward()
            grad_norm = apply_defense_to_gradients(model, condition.defense)
            scaler.step(optimizer)
            scaler.update()

            with torch.no_grad():
                batch_acc = accuracy_from_logits(logits, y)
                preds = torch.argmax(logits, dim=1)

            losses.append(float(loss.item()))
            accs.append(batch_acc)
            all_preds.append(preds.detach().cpu().numpy())
            all_labels.append(y.detach().cpu().numpy())
            all_protected.append(s.detach().cpu().numpy())

            if (batch_idx == 1) or (batch_idx % cfg.log_batch_every == 0):
                logger.info(
                    f"[client={client_id}] epoch={epoch} batch={batch_idx} "
                    f"loss={loss.item():.5f} acc={batch_acc:.4f} grad_norm={grad_norm:.4f} "
                    f"attack={condition.attack} defense={condition.defense}"
                )

    preds_np = np.concatenate(all_preds) if all_preds else np.array([], dtype=np.int64)
    labels_np = np.concatenate(all_labels) if all_labels else np.array([], dtype=np.int64)
    prot_np = np.concatenate(all_protected) if all_protected else np.array([], dtype=np.int64)

    dp_gap = demographic_parity_gap(preds_np, prot_np) if preds_np.size else 0.0
    eo_gap = equalized_odds_gap(preds_np, labels_np, prot_np) if preds_np.size else 0.0

    local_state_cpu = {k: v.detach().cpu() for k, v in model.state_dict().items()}
    duration = time.perf_counter() - t0

    metrics = ClientMetrics(
        client_id=client_id,
        loss=float(np.mean(losses)) if losses else 0.0,
        accuracy=float(np.mean(accs)) if accs else 0.0,
        demographic_parity_gap=float(dp_gap),
        equalized_odds_gap=float(eo_gap),
        samples=len(subset),
        duration_sec=float(duration),
    )

    logger.info(
        f"[client={client_id}] done | samples={metrics.samples} loss={metrics.loss:.5f} "
        f"acc={metrics.accuracy:.4f} dp_gap={metrics.demographic_parity_gap:.4f} "
        f"eo_gap={metrics.equalized_odds_gap:.4f} duration={format_seconds(metrics.duration_sec)}"
    )

    return local_state_cpu, metrics


def fedavg(client_states: List[Dict[str, torch.Tensor]], client_weights: List[float]) -> Dict[str, torch.Tensor]:
    if not client_states:
        raise ValueError("fedavg requires at least one client state")

    w = np.array(client_weights, dtype=np.float64)
    w = w / np.clip(w.sum(), 1e-12, None)

    avg_state: Dict[str, torch.Tensor] = {}
    keys = client_states[0].keys()

    for k in keys:
        acc = None
        for i, st in enumerate(client_states):
            term = st[k].float() * float(w[i])
            acc = term if acc is None else acc + term
        avg_state[k] = acc

    return avg_state


@torch.no_grad()
def evaluate_global_model(model: nn.Module, val_loader: DataLoader, device: torch.device, logger: logging.Logger) -> Dict[str, float]:
    model.eval()
    criterion = nn.CrossEntropyLoss()

    losses: List[float] = []
    all_preds: List[np.ndarray] = []
    all_labels: List[np.ndarray] = []
    all_protected: List[np.ndarray] = []

    for batch_idx, (x, y, s) in enumerate(val_loader, start=1):
        x = x.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)
        s = s.to(device, non_blocking=True)

        logits = model(x)
        loss = criterion(logits, y)
        preds = torch.argmax(logits, dim=1)

        losses.append(float(loss.item()))
        all_preds.append(preds.detach().cpu().numpy())
        all_labels.append(y.detach().cpu().numpy())
        all_protected.append(s.detach().cpu().numpy())

        if batch_idx == 1:
            logger.info(f"[eval] first batch loss={loss.item():.5f} acc={accuracy_from_logits(logits, y):.4f}")

    preds_np = np.concatenate(all_preds) if all_preds else np.array([], dtype=np.int64)
    labels_np = np.concatenate(all_labels) if all_labels else np.array([], dtype=np.int64)
    prot_np = np.concatenate(all_protected) if all_protected else np.array([], dtype=np.int64)

    val_acc = float((preds_np == labels_np).mean()) if preds_np.size else 0.0
    val_dp = demographic_parity_gap(preds_np, prot_np) if preds_np.size else 0.0
    val_eo = equalized_odds_gap(preds_np, labels_np, prot_np) if preds_np.size else 0.0

    return {
        "val_loss": float(np.mean(losses)) if losses else 0.0,
        "val_accuracy": val_acc,
        "val_demographic_parity_gap": float(val_dp),
        "val_equalized_odds_gap": float(val_eo),
    }


def run_single_condition(
    cfg: ExperimentConfig,
    logger: logging.Logger,
    run_name: str,
    condition_name: str,
    train_dataset: Dataset,
    val_dataset: Dataset,
    client_partitions: Dict[int, List[int]],
    deadline_ts: float,
) -> ConditionRunResult:
    if condition_name not in CONDITION_REGISTRY:
        raise KeyError(f"Unknown condition: {condition_name}")

    condition = CONDITION_REGISTRY[condition_name]
    devices = get_available_devices()
    primary_device = devices[0]

    logger.info(
        f"=== Condition {condition.name} | attack={condition.attack} defense={condition.defense} "
        f"| devices={devices} ==="
    )

    global_model = build_model(num_classes=2, device=primary_device, logger=logger)
    global_state = {k: v.detach().cpu() for k, v in global_model.state_dict().items()}

    checkpoint_mgr = CheckpointManager(cfg.checkpoints_root / run_name, condition_name, logger)
    history: List[Dict[str, Any]] = []
    start_round = 1

    if cfg.resume_from_checkpoint:
        ckpt = checkpoint_mgr.load_latest()
        if ckpt is not None:
            global_state = ckpt.global_model_state
            history = ckpt.history
            start_round = ckpt.round_idx + 1
            logger.info(f"Resuming condition={condition_name} from round {start_round}")

    eta = AdaptiveETAEstimator(alpha=0.30)
    val_loader = DataLoader(
        val_dataset,
        batch_size=cfg.batch_size,
        shuffle=False,
        num_workers=cfg.num_workers,
        pin_memory=(cfg.pin_memory and primary_device.type == "cuda"),
    )

    wallclock_start = time.time()
    stopped_reason = "completed"

    for round_idx in range(start_round, cfg.rounds + 1):
        if time.time() >= deadline_ts:
            stopped_reason = "runtime_budget_reached"
            logger.warning("Runtime budget reached, stopping cleanly after checkpoint.")
            break

        round_t0 = time.perf_counter()
        remaining = cfg.rounds - round_idx

        rng = np.random.default_rng(SEED + round_idx)
        selected_clients = rng.choice(np.arange(cfg.num_clients), size=cfg.clients_per_round, replace=False).tolist()
        logger.info(
            f"[round={round_idx}/{cfg.rounds}] selected_clients={selected_clients} "
            f"remaining_rounds={remaining}"
        )

        local_states: List[Dict[str, torch.Tensor]] = []
        local_weights: List[float] = []
        local_metrics: List[ClientMetrics] = []

        for i, cid in enumerate(selected_clients):
            device = devices[i % len(devices)]
            st, cm = local_train_one_client(
                client_id=cid,
                device=device,
                global_state=global_state,
                train_dataset=train_dataset,
                sample_indices=client_partitions[cid],
                cfg=cfg,
                condition=condition,
                logger=logger,
            )
            local_states.append(st)
            local_weights.append(float(max(1, cm.samples)))
            local_metrics.append(cm)

        global_state = fedavg(local_states, local_weights)
        global_model.load_state_dict({k: v.to(primary_device) for k, v in global_state.items()})

        eval_metrics = evaluate_global_model(global_model, val_loader, primary_device, logger)

        train_loss = float(np.mean([m.loss for m in local_metrics])) if local_metrics else 0.0
        train_acc = float(np.mean([m.accuracy for m in local_metrics])) if local_metrics else 0.0

        round_dur = time.perf_counter() - round_t0
        eta.update(round_dur)
        eta_sec, eta_low, eta_high = eta.estimate(cfg.rounds - round_idx)
        elapsed_sec = time.time() - wallclock_start

        rm = RoundMetrics(
            round_idx=round_idx,
            selected_clients=selected_clients,
            aggregate_train_loss=train_loss,
            aggregate_train_accuracy=train_acc,
            val_loss=eval_metrics["val_loss"],
            val_accuracy=eval_metrics["val_accuracy"],
            val_demographic_parity_gap=eval_metrics["val_demographic_parity_gap"],
            val_equalized_odds_gap=eval_metrics["val_equalized_odds_gap"],
            round_duration_sec=float(round_dur),
            elapsed_sec=float(elapsed_sec),
            eta_sec=float(eta_sec),
            eta_low_sec=float(eta_low),
            eta_high_sec=float(eta_high),
        )

        rm_dict = asdict(rm)
        history.append(rm_dict)
        checkpoint_mgr.append_round_metrics(rm_dict)

        logger.info(
            f"[round={round_idx}] train_loss={rm.aggregate_train_loss:.5f} train_acc={rm.aggregate_train_accuracy:.4f} "
            f"val_loss={rm.val_loss:.5f} val_acc={rm.val_accuracy:.4f} "
            f"val_dp={rm.val_demographic_parity_gap:.4f} val_eo={rm.val_equalized_odds_gap:.4f} "
            f"round_time={format_seconds(rm.round_duration_sec)} "
            f"ETA={format_seconds(rm.eta_sec)} [{format_seconds(rm.eta_low_sec)}..{format_seconds(rm.eta_high_sec)}]"
        )

        if (round_idx % cfg.checkpoint_every_round) == 0:
            checkpoint_mgr.save(
                round_idx=round_idx,
                condition_name=condition_name,
                global_model_state=global_state,
                history=history,
                wallclock_elapsed_sec=elapsed_sec,
            )

    finished_round = history[-1]["round_idx"] if history else 0

    return ConditionRunResult(
        condition_name=condition_name,
        started_from_round=start_round,
        finished_round=finished_round,
        stopped_reason=stopped_reason,
        total_duration_sec=float(time.time() - wallclock_start),
        rounds=history,
    )


def save_json(path: Path, payload: Dict[str, Any]) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open("w", encoding="utf-8") as f:
        json.dump(payload, f, indent=2)


def run_full_experiment(cfg: ExperimentConfig, logger: logging.Logger) -> Dict[str, Any]:
    run_name = datetime.now().strftime("full_run_%Y%m%d_%H%M%S")
    logger.info(f"Starting FULL RUN: {run_name}")

    train_dataset, val_dataset, client_partitions, dataset_source = build_datasets_and_partitions(cfg, logger)

    deadline_ts = time.time() + cfg.runtime_budget_seconds()
    all_results: List[Dict[str, Any]] = []

    for condition_name in cfg.condition_names:
        if time.time() >= deadline_ts:
            logger.warning("Global runtime budget reached before next condition.")
            break

        if condition_name not in CONDITION_REGISTRY:
            logger.warning(f"Skipping unknown condition: {condition_name}")
            continue

        result = run_single_condition(
            cfg=cfg,
            logger=logger,
            run_name=run_name,
            condition_name=condition_name,
            train_dataset=train_dataset,
            val_dataset=val_dataset,
            client_partitions=client_partitions,
            deadline_ts=deadline_ts,
        )
        all_results.append(asdict(result))

    summary = {
        "mode": "full_run",
        "run_name": run_name,
        "dataset_source": dataset_source,
        "config": {k: (str(v) if isinstance(v, Path) else v) for k, v in asdict(cfg).items()},
        "conditions": all_results,
        "finished_at": datetime.now(timezone.utc).isoformat(),
    }

    json_path = cfg.results_root / f"{run_name}.json"
    save_json(json_path, summary)

    # Flat CSV export for quick plotting.
    flat_rows: List[Dict[str, Any]] = []
    for cond in all_results:
        for r in cond.get("rounds", []):
            row = {"condition_name": cond["condition_name"], **r}
            flat_rows.append(row)
    if flat_rows:
        csv_path = cfg.results_root / f"{run_name}_rounds.csv"
        pd.DataFrame(flat_rows).to_csv(csv_path, index=False)
        logger.info(f"Saved per-round CSV -> {csv_path}")

    logger.info(f"Saved full run summary -> {json_path}")
    return summary


def run_dry_run_suite(cfg: ExperimentConfig, logger: logging.Logger) -> Dict[str, Any]:
    run_name = datetime.now().strftime("dry_run_%Y%m%d_%H%M%S")
    logger.info(f"Starting DRY RUN: {run_name}")

    checks: List[Dict[str, Any]] = []

    def run_check(name: str, fn):
        t0 = time.perf_counter()
        try:
            fn()
            checks.append({
                "name": name,
                "status": "pass",
                "duration_sec": time.perf_counter() - t0,
                "error": "",
            })
            logger.info(f"[dry-check] PASS {name}")
        except Exception as ex:
            checks.append({
                "name": name,
                "status": "fail",
                "duration_sec": time.perf_counter() - t0,
                "error": f"{type(ex).__name__}: {ex}",
            })
            logger.exception(f"[dry-check] FAIL {name}")

    cache: Dict[str, Any] = {}

    run_check("filesystem_paths", lambda: [p.mkdir(parents=True, exist_ok=True) for p in [cfg.artifacts_root, cfg.checkpoints_root, cfg.logs_root, cfg.results_root]])

    def _dataset_check():
        train_dataset, val_dataset, client_partitions, dataset_source = build_datasets_and_partitions(cfg, logger)
        cache["train_dataset"] = train_dataset
        cache["val_dataset"] = val_dataset
        cache["client_partitions"] = client_partitions
        cache["dataset_source"] = dataset_source

    run_check("dataset_partition_loading", _dataset_check)

    def _model_forward_backward_check():
        devices = get_available_devices()
        device = devices[0]
        model = build_model(num_classes=2, device=device, logger=logger)
        model.train()
        x = torch.randn(8, 3, cfg.image_size, cfg.image_size, device=device)
        y = torch.randint(low=0, high=2, size=(8,), device=device)
        criterion = nn.CrossEntropyLoss()
        out = model(x)
        loss = criterion(out, y)
        loss.backward()

    run_check("model_forward_backward", _model_forward_backward_check)

    def _checkpoint_cycle_check():
        logger_local = logger
        cpm = CheckpointManager(cfg.checkpoints_root / run_name, "dry_checkpoint_test", logger_local)
        dummy_model = {"w": torch.randn(4, 4)}
        cpm.save(1, "dry_checkpoint_test", dummy_model, history=[{"round_idx": 1}], wallclock_elapsed_sec=1.23)
        loaded = cpm.load_latest()
        assert loaded is not None
        assert loaded.round_idx == 1

    run_check("checkpoint_save_load", _checkpoint_cycle_check)

    def _mini_fl_check():
        mini_cfg = deepcopy(cfg)
        mini_cfg.condition_names = ["vanilla", "A1_only"]
        mini_cfg.rounds = min(cfg.rounds, 2)
        mini_cfg.num_clients = max(2, cfg.num_clients)
        mini_cfg.clients_per_round = mini_cfg.num_clients
        mini_cfg.max_batches_per_epoch = min(cfg.max_batches_per_epoch, 2)
        mini_cfg.batch_size = min(cfg.batch_size, 16)
        mini_cfg.runtime_buffer_minutes = 0
        mini_cfg.runtime_budget_hours = 0.10

        train_dataset = cache["train_dataset"]
        val_dataset = cache["val_dataset"]
        client_partitions = cache["client_partitions"]

        run_local_name = f"{run_name}_mini"
        deadline_ts = time.time() + mini_cfg.runtime_budget_seconds()

        for cond_name in mini_cfg.condition_names:
            _ = run_single_condition(
                cfg=mini_cfg,
                logger=logger,
                run_name=run_local_name,
                condition_name=cond_name,
                train_dataset=train_dataset,
                val_dataset=val_dataset,
                client_partitions=client_partitions,
                deadline_ts=deadline_ts,
            )

    run_check("mini_federated_roundtrip", _mini_fl_check)

    failed = [c for c in checks if c["status"] == "fail"]
    status = "success" if not failed else "failed"

    report = {
        "mode": "dry_run",
        "run_name": run_name,
        "status": status,
        "checks": checks,
        "dataset_source": cache.get("dataset_source", "unknown"),
        "config": {k: (str(v) if isinstance(v, Path) else v) for k, v in asdict(cfg).items()},
        "finished_at": datetime.now(timezone.utc).isoformat(),
    }

    report_path = cfg.results_root / f"{run_name}_report.json"
    save_json(report_path, report)
    logger.info(f"Saved dry-run report -> {report_path}")

    if failed:
        raise RuntimeError(f"Dry run failed with {len(failed)} failed checks")

    return report


print("Federated pipeline functions are ready.")
```

```python
# ==== End-to-end execution cell ====
set_all_seeds(SEED)

run_mode = "dry_run" if CFG.dry_run else "full_run"
run_stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
logger = build_run_logger(CFG.logs_root, f"{run_mode}_{run_stamp}")

logger.info("============================================================")
logger.info(f"Starting mode={run_mode}")
logger.info(f"Runtime budget={format_seconds(CFG.runtime_budget_seconds())}")
logger.info(f"Conditions={CFG.condition_names}")
logger.info("============================================================")

if CFG.dry_run:
    dry_report = run_dry_run_suite(CFG, logger)
    print("Dry run completed successfully.")
    print(json.dumps({"status": dry_report["status"], "run_name": dry_report["run_name"]}, indent=2))
else:
    full_summary = run_full_experiment(CFG, logger)
    print("Full run completed (or stopped at budget).")
    print(json.dumps({
        "run_name": full_summary["run_name"],
        "conditions_executed": [c["condition_name"] for c in full_summary["conditions"]],
    }, indent=2))
```

```python
def configure_for_two_hour_budget(cfg: ExperimentConfig) -> ExperimentConfig:
    cfg = deepcopy(cfg)
    cfg.runtime_budget_hours = 2.0
    cfg.runtime_buffer_minutes = 5

    # Conservative defaults for dual L40S to stay inside budget.
    cfg.num_clients = min(cfg.num_clients, 6)
    cfg.clients_per_round = min(cfg.clients_per_round, 3)
    cfg.rounds = min(cfg.rounds, 30)
    cfg.local_epochs = min(cfg.local_epochs, 1)
    cfg.max_batches_per_epoch = min(cfg.max_batches_per_epoch, 40)

    return cfg


def configure_condition_preset(cfg: ExperimentConfig, preset: str = "fast_a1") -> ExperimentConfig:
    cfg = deepcopy(cfg)

    presets = {
        "fast_a1": ["vanilla", "A1_only"],
        "a1_defenses": ["vanilla", "A1_only", "A1_D1", "A1_D2", "A1_D3"],
        "exploratory": ["vanilla", "A1_only", "A2_only", "A3_only"],
    }

    if preset not in presets:
        raise ValueError(f"Unknown preset={preset}. Available: {list(presets)}")

    cfg.condition_names = presets[preset]
    return cfg


def launch_with_preset(
    dry_run: bool,
    full_run: bool,
    resume: bool = True,
    preset: str = "fast_a1",
) -> Dict[str, Any]:
    cfg = build_config(dry_run=dry_run, full_run=full_run, resume_from_checkpoint=resume)
    cfg = configure_for_two_hour_budget(cfg)
    cfg = configure_condition_preset(cfg, preset=preset)

    run_mode = "dry_run" if cfg.dry_run else "full_run"
    run_stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    logger = build_run_logger(cfg.logs_root, f"{run_mode}_{run_stamp}_preset_{preset}")

    logger.info(f"Launching with preset={preset} mode={run_mode}")
    if cfg.dry_run:
        return run_dry_run_suite(cfg, logger)
    return run_full_experiment(cfg, logger)


print("Reusable preset launchers are ready.")
print("Example usage:")
print("  launch_with_preset(dry_run=True, full_run=False, preset='fast_a1')")
print("  launch_with_preset(dry_run=False, full_run=True, preset='a1_defenses')")
```
